```python
# functions (3)
import numpy as np

def normalize(lst):
    zlst = np.empty(len(lst))
    mean = np.mean(lst)
    std = np.std(lst)
    for idx, x in enumerate(lst):
        zlst[idx] = (x - mean)/std
    return zlst

print(normalize(np.array([1,3,5,7,9])))
```

```python
# default setting for plotting and loading the data
import pandas as pd  
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load the data
totals = pd.read_csv("https://piazza.com/class_profile/get_resource/ij0cpi5tuaa4wv/ik6hlfvu8dk1vu", delimiter='\t', index_col=0)
guns = pd.read_csv("https://piazza.com/class_profile/get_resource/ij0cpi5tuaa4wv/ik6hlau5u2i26d", delimiter='\t', index_col=0)
gdp = pd.read_csv("https://piazza.com/class_profile/get_resource/ij0cpi5tuaa4wv/ik6hl6z36e81t3", delimiter='\t', index_col=1)

# Check each of the data set

# join the gun homicides and GDP data by country 
data = totals.join(guns).join(gdp)

# Add a column 'Gun Percent', that calculatees the percentage of gun homicides among all of homicides
data['Gun Percent'] = 100 * data['Gun Homicides'] / data['Homicides']

# Delete 'Unintentional', 'Undetermined' and 'Gun Suicides' columns from the data
del data['Unintentional'],data['Undetermined'],data['Gun Suicides']

# Drop the N/A values 
data = data.dropna()
data
```

```python
# Which countries are included in the data set?

for country in data.index:
    print(country)
```

```python
# Make a function that takes a data set and a number (default value 10) 
# and plots the percentage of gun homicide
#   The y-axis value range: 0 ~ 100
#   The title of the plot: "% Gun Homicide"

def plot_percent(df, limit=10):
    df['Gun Percent'][:limit].plot()
    plt.ylim(0,100)
    plt.title("% Gun Homicide")
    plt.show()
```

```python
# Make a function that takes a data set and a number (default value 10) 
# and shows the data table that includes 'Homicies', 'Gun Homicies' 
# and 'Gun Data Source' columns

def display_relevant(df, limit=10):
    print(df.ix[:,['Homicides', 'Gun Homicides', 'Gun Data Source']][:limit])
```

```python
# Sort the data by the descending order of gun homicides 
# and display a table of the top 15 countries
# Also, plot the % gun homicide of those countries 
data = data.sort("Gun Homicides", ascending=False)
display_relevant(data, 15)
plot_percent(data, 15)
```

```python
# Create a function that shows a stacked bar plot of
# gun homicides and total homicides

def plot_stacked(df, limit=10):
    df = df.ix[:,['Homicides', 'Gun Homicides']][:limit]
    df['Total Homicides'] = df['Homicides'] - df['Gun Homicides']
    del df['Homicides']
    df.plot(kind='bar', stacked=True, sort_columns=True)
    plt.ylabel("per 100k")
    plt.show()
```

```python
# Sort the data by the descending order of gun homicides 
# and display a table of the top 15 countries
# Also, plot the % gun homicide of those countries 


data = data.sort("Gun Homicides", ascending=False)
display_relevant(data, 15)
plot_percent(data, 15)
```

```python
# Store the top 30 richest countries
top = data.sort('GDP')[-30:]

# Sort by the number of gun homicides
top_by_guns = top.sort("Gun Homicides", ascending=False)

# print the data table and
display_relevant(top_by_guns, 5)

# draw a stacked bar graph
plot_stacked(top_by_guns, 10)
```

```python
# Select the European countries, the United States, Canada
# Mexico for data countries
index = (data['Region'] == 'Europe') + \
        (data.index == 'United States') + \
        (data.index == 'Canada') + \
        (data.index == 'Mexico')
selected = data[index]

# Sort by the number of gun homicides
by_guns = selected.sort("Gun Homicides", ascending=False)


# Plot the gun homicides rates of the top 25 countries and
plot_stacked(by_guns, limit=25)


# the data of the countries
display_relevant(selected, limit=None)
```
